```python
# Import Libraries
from stanfordcorenlp import StanfordCoreNLP
import json
import os
import re

nlp = StanfordCoreNLP('./stanford-corenlp-full-2018-10-05')
```

```python
import numpy as np
import pandas as pd
import os, itertools, csv
from bs4 import BeautifulSoup
import re
from pytrends.request import TrendReq
from datetime import *
import time
import nltk

# load the dataset
dataset = pd.read_csv('./data/test.csv')
dataset.head()
```

```python
def get_article(text: str) -> str:    
    soup = BeautifulSoup(text, "html.parser")
    article = soup.find('section',{'class':'article-content'})
    return article.get_text()

res = []

for i in range(dataset.shape[0]):
    res.append(get_article(dataset['Page content'][i]))
dataset['article'] = res
```

```python
dataset.head()
```

```python
from nltk.tokenize import sent_tokenize

veryNegative = []
negative = []
neutral = []
positive = []
veryPositive = []
invalid = []
articles = dataset['article'].values
for i in range(dataset.shape[0]): 
    sentence = sent_tokenize(articles[i])
    print(i, "Finish:", i/len(articles))
    # Init variables
    veryNegativeResults = 0
    negativeResults = 0
    neutralResults = 0
    positiveResults = 0
    veryPositiveResults = 0
    invalidResults = 0
    for sen in sentence:
        try:
            jsonSentiment = json.loads(nlp.annotate(sen,
                                            properties={
                                                'annotators': 'sentiment',
                                                'outputFormat': 'json'
                                            }))
            currentSentence = ''        
            for s in jsonSentiment["sentences"]:
                for y in s["tokens"]:
                    currentSentence = currentSentence + ' ' +  y["originalText"]
                    #print(currentSentence)
                if(s["sentiment"] == "Verynegative"):
                    veryNegativeResults = veryNegativeResults + 1
                elif(s["sentiment"] == "Negative"):
                    negativeResults = negativeResults + 1
                elif(s["sentiment"] == "Neutral"):
                    neutralResults = neutralResults + 1
                elif(s["sentiment"] == "Positive"):
                    positiveResults = positiveResults + 1
                elif(s["sentiment"] == "Verypositive"):
                    veryPositiveResults = veryPositiveResults + 1
                else:
                    invalidResults = invalidResults + 1
        except:
            print("Error:", sen)
    veryNegative.append(veryNegativeResults/len(sentence))
    negative.append(negativeResults/len(sentence))
    neutral.append(neutralResults/len(sentence))
    positive.append(positiveResults/len(sentence))
    veryPositive.append(veryPositiveResults/len(sentence))
    invalid.append(invalidResults/len(sentence))
```
